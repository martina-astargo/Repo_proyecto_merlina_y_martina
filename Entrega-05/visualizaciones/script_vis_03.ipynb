```python
import altair as alt

source = "/content/database_limpia_intento2.xlsx"

gaussian_jitter = alt.Chart(source, title='Asociaciones de género').mark_circle(size=8).encode(
    y="Formas que predominan:N",
    x="Años:Q", # Changed G to Q
    yOffset="Género:N", # Changed B to N
    color=alt.Color('Major Genre:N').legend(None)
).transform_calculate(
    # Generate Gaussian jitter with a Box-Muller transform
    Género="sqrt(-2*log(random()))*cos(2*PI*random())"
)

uniform_jitter = gaussian_jitter.transform_calculate(
    # Generate uniform jitter
    Género='random()'
).encode(
    alt.Y('Major Genre:N').axis(None)
).properties(
    title='Asociaciones de género'
)

(gaussian_jitter | uniform_jitter).resolve_scale(yOffset='independent')
```

```python
import altair as alt
import pandas as pd

# Cargar tus datos
df = pd.read_excel('/content/database_limpia_intento2.xlsx')

# Definir la columna 'agrupaciones' con los valores deseados
def map_roles_to_agrupaciones(role):
    if role == 'Protagonista':
        return 'dominante'
    elif role == 'Antagonista':
        return 'conflictivo'
    elif role == 'Secundario':
        return 'emotivo'
    else:
        return 'Otro' # Handles any other roles not explicitly mapped

df['agrupaciones'] = df['roles'].apply(map_roles_to_agrupaciones)

chart = alt.Chart(df).mark_circle(size=60).encode(
    x=alt.X('año:O', title='Año'),
    y=alt.Y('agrupaciones:N', title='Impulso dramático'),
    color=alt.Color('genero:N', title='Género', scale=alt.Scale(range=['red', 'blue', 'green', 'orange'])), # Custom colors added here
    tooltip=['personaje', 'genero', 'agrupaciones', 'año']
).properties(
    title='Distribución de impulsos dramáticos por año y género'
).transform_calculate(
    # aplicar jitter para dispersión visual
    jitter="sqrt(-2*log(random()))*cos(2*PI*random())"
).encode(
    yOffset='jitter:Q'
)

chart
```

```python
import pandas as pd
import altair as alt

df = pd.read_excel('/content/database_limpia_intento2.xlsx')

df['Agrupaciones'] = df['Agrupaciones'].astype(str).str.strip().str.lower()
df['Agrupaciones'] = df['Agrupaciones'].str.replace(r'\s+', ' ', regex=True)
df['Agrupaciones'] = df['Agrupaciones'].str.title()

chart = alt.Chart(df).mark_circle(size=80).encode(
    x=alt.X('año:O', title='Año'),
    y=alt.Y('genero:N', title='Género'),
    color=alt.Color(
        'Agrupaciones:N',
        title='Tipo de agrupación',
        scale=alt.Scale(domain=['Conflictivo', 'Dominante', 'Emotivo', 'Otro'], range=['red', 'blue', 'yellow', 'gray']) # Custom colors applied here
    ),
    tooltip=['personaje', 'genero', 'Agrupaciones', 'año']
).properties(
    title='Relación entre género y la fuerza que actúa' # Title changed here
).transform_calculate(
    jitter="sqrt(-2*log(random()))*cos(2*PI*random())"
).encode(
    xOffset='jitter:Q'
)

chart
```
